```python
import requests
import bs4
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
```

```python
url_inicial = 'https://estudiosdemograficosyurbanos.colmex.mx/index.php/edu/issue/archive'
url_root = 'https://estudiosdemograficosyurbanos.colmex.mx/index.php/edu/issue/archive'
r=requests.get(url_inicial)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
box = soup.find('ul', class_='issues_archive')
volumen=box.findAll('div', class_='obj_issue_summary')
vol = [x.find('a').get('href')for x in volumen]
vol=[urljoin(url_root,i) for i in vol]
vol
```

```python
def get_url_items(sopa, url): 
    box = soup.find('ul', class_='issues_archive')
    volumen=box.findAll('div', class_='obj_issue_summary')
    vol = [x.find('a').get('href')for x in volumen]
    vol=[urljoin(url_root,i) for i in vol]
    return vol
```

```python
links_items=[]
i=0
while i<1:
    i+=1
    print(f'Estoy en la pagina {url_inicial}')
    r_pag=requests.get(url_inicial)
    s_p=BeautifulSoup(r_pag.text,'html.parser')
    links=get_url_items(s_p, url_inicial)
    links_items.append(links) 
```

```python
list_scraper=[]
for i in links_items:
    for j in i:
        list_scraper.append(j)
len(list_scraper)
```

```python
uno=list_scraper[0]
r_item=requests.get(uno)
s_item=BeautifulSoup(r_item.text, 'html.parser')
```

```python
#funcion para iniciar el escraper de cada libro
def scraper_book(url):
    content_book={}
    r=requests.get(url)
    s_item=BeautifulSoup(r.text,'html.parser')
    #volumen del libro
    try:
        vol=s_item.find('h1').get_text(strip=True)
        content_book['Volumen']=vol
    except AttributeError:
        content_book['Volumen']=None
    #articulo del libro
    try:
        arti=s_item.find('h3').get_text(strip=True)
        content_book['Articulo']=arti
    except AttributeError:
        content_book['Articulo']=None
    #autor del libro
    try:
        autor=s_item.find('div', class_='authors').get_text(strip=True)
        content_book['Autor']=autor
    except AttributeError:
        content_book['Autor']=None
    #obtener link
    try:
        link=s_item.find('a', class_='obj_galley_link pdf').get('href')
        content_book['Link']=urljoin(url_root, link)
    except AttributeError:
        content_book['Link']=None
    return content_book
```

```python
list_scraper=list_scraper[0:168]
datos_book=[]
for idx, i in enumerate(list_scraper):
    print(f'estas escrapeando la pag {idx}')
    datos_book.append(scraper_book(i))
```

```python
df_catalogo=pd.DataFrame(datos_book)
df_catalogo
```

```python
df_catalogo.to_csv('Revista33.csv')
```
