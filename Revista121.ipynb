```python
import requests
import bs4
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
```

```python
url_inicial = 'https://www.redalyc.org/revista.oa?id=612&numero=58143'
url_root = 'https://www.redalyc.org/revista.oa?id=612&numero=58143'
r=requests.get(url_inicial)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
#box = soup.find('div', class_='ui-accordion-content ui-helper-reset ui-widget-content ui-corner-bottom ui-accordion-content-active')
volumen=soup.findAll('div', id='lista-articulos')
vol = [x.find('a').get('href')for x in volumen]
vol=[urljoin(url_root,i) for i in vol]
vol
```

```python
def get_url_items(sopa,url):
    #box1 = soup1.find('div', class_='sections') 
    volumen=soup.findAll('div', id='lista-articulos')
    vol = [x.find('a').get('href')for x in volumen]
    return vol
```

```python
links_items=[]
i=0
while i<1:
    i+=1
    print(f'Estoy en la pagina {url_inicial}')
    r_pag=requests.get(url_inicial)
    s_p=BeautifulSoup(r_pag.text,'html.parser')
    links=get_url_items(s_p, url_inicial)
    links_items.append(links)
```

```python
list_scraper=[]
for i in links_items:
    for j in i:
        list_scraper.append(j)
len(list_scraper)
```

```python
uno=list_scraper[0]
r_item=requests.get(uno)
s_item=BeautifulSoup(r_item.text, 'html.parser')
```

```python
#funcion para iniciar el escraper de cada libro
def scraper_book(url):
    content_book={}
    r=requests.get(url)
    tire='Revista Mexicana de Fitopatología'
    a='VI. Biotecnología y Ciencias Agropecuarias'
    tem='Etiología, Manejo Fitosanitario, Epidemiología, Biología, Taxonomía, Control de Hongos, Bacterias, Virus, Nematodos y Otros Organismos Fitopatógenos'
    s_item=BeautifulSoup(r.text,'html.parser')
    #titulo de revista
    try:
        titulo=tire
        content_book['Titulo Revista']=tire
    except AttributeError:
        content_book['Titulo Revista']=None
    #area
    try:
        area=a
        content_book['Area']=a
    except AttributeError:
        content_book['Area']=None
    #tematica
    try:
        tema=tem
        content_book['Tematica']=tem
    except AttributeError:
        content_book['Tematica']=None
    #titulo articulo
    try:
        titu=s_item.find('span', id='titulo-articulo').get_text(strip=True)
        content_book['Titulo Articulo']=titu
    except AttributeError:
        content_book['Titulo Articulo']=None
    #resumen
    try:
        resu=s_item.find('span', class_='capital').get_text(strip=True)
        content_book['Resumen']=resu.replace('Resumen.:','')
    except AttributeError:
        content_book['Resumen']=None
    #abstract
    try:
        abst=s_item.find('p', class_='articulo-resumen-traduccion').get_text(strip=True)
        content_book['Abstract']=abst.replace('ABSTRACT:','')
    except AttributeError:
        content_book['Abstract']=None
    #Link incial
    try:
        linkart=url
        content_book['Link Articulo']=linkart
    except AttributeError:
        content_book['Link Articulo']=None
    #link articulo
    try:
        link=s_item.find('div', title="PDF").find('a').get('href')
        content_book['Link PDF']=urljoin(url_root, link)
    except AttributeError:
        content_book['Link PDF']=None
    return content_book
```

```python
list_scraper=list_scraper[0:12]
datos_book=[]
for idx, i in enumerate(list_scraper):
    print(f'estas escrapeando la pag {idx}')
    datos_book.append(scraper_book(i))
```

```python
df_catalogo=pd.DataFrame(datos_book)
df_catalogo
```
