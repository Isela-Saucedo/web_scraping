```python
import requests
import bs4
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
```

```python
url_inicial = 'https://revistas.chapingo.mx/forestales/?section=articles&subsec=issues'
url_root = 'https://revistas.chapingo.mx/forestales/?section=articles&subsec=issues'
r=requests.get(url_inicial)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
box = soup.find('div', id='accordionExample')
volumen=box.findAll('div', class_='card')
vol = [x.find('a').get('href')for x in volumen]
vol=[urljoin(url_root,i) for i in vol]
vol
```

```python
def get_url_items(sopa, url): 
    box = soup.find('div', id='accordionExample')
    volumen=box.findAll('div', class_='card')
    vol = [x.find('a').get('href')for x in volumen]
    vol=[urljoin(url_root,i) for i in vol]
    return vol
```

```python
links_items=[]
i=0
while i<1:
    i+=1
    print(f'Estoy en la pagina {url_inicial}')
    r_pag=requests.get(url_inicial)
    s_p=BeautifulSoup(r_pag.text,'html.parser')
    links=get_url_items(s_p, url_inicial)
    links_items.append(links) 
    #next_a=s_p.select('div.cmp_pagination > a')
    #if not next_a or not next_a[0].get('href'):
     #   break
    #url_inicial = urljoin(url_inicial, next_a[0].get('href'))
```

```python
list_scraper=[]
for i in links_items:
    for j in i:
        list_scraper.append(j)
len(list_scraper)
```

```python
uno=list_scraper[0]
r_item=requests.get(uno)
s_item=BeautifulSoup(r_item.text, 'html.parser')
```

```python
#funcion para iniciar el escraper de cada libro
def scraper_book(url):
    content_book={}
    r=requests.get(url)
    tire='Revista Chapingo Serie Ciencias Forestales y del Ambiente'
    a='VI. Biotecnología y Ciencias Agropecuarias'
    tem='Silvicultura y Manejo, Estadística Computo y Matemáticas, Ambiental, Economía y Sociología Forestal'
    url='http://www.chapingo.mx/revistas/forestales'
    s_item=BeautifulSoup(r.text,'html.parser')
    #titulo de revista
    try:
        titulo=tire
        content_book['Titulo de revista']=tire
    except AttributeError:
        content_book['Titulo de revista']=None
    #area
    try:
        area=a
        content_book['Área']=a
    except AttributeError:
        content_book['Área']=None
    #tematica
    try:
        tema=tem
        content_book['Temática']=tem
    except AttributeError:
        content_book['Temática']=None
    #titulo libro
    try:
        titu=s_item.find('h1', class_='titulos pb-3').get_text(strip=True)
        content_book['Titulo del libro']=titu
    except AttributeError:
        content_book['Titulo del libro']=None
    #resumen
    try:
        resu=s_item.find('div', id='articleAbstract').get_text(strip=True)
        content_book['Resumen']=resu
    except AttributeError:
        content_book['Resumen']=None
    #Link incial
    try:
        inicial=url
        content_book['Link inicial']=url
    except AttributeError:
        content_book['Link inicial']=None
    #link articulo
    try:
        link=s_item.find('a', class_='link-art pr-3').get('href')
        content_book['Link articulo']=urljoin(url_root, link)
    except AttributeError:
        content_book['Link articulo']=None
    return content_book
```

```python
list_scraper=list_scraper[0:25]
datos_book=[]
for idx, i in enumerate(list_scraper):
    print(f'estas escrapeando la pag {idx}')
    datos_book.append(scraper_book(i))
```

```python
df_catalogo=pd.DataFrame(datos_book)
df_catalogo
```
